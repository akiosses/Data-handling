```python
#separating the rows the rows that have multiple SNPs,RA,ChrId,ChrPos which are separated by    , or ; or / or x 

data=pd.read_csv("cleanGWAS_and_beta_16_04_21.csv")
data=data[data["SNP"].astype(str).str.contains("rs")]
data["publication_year"]=data["publication_year"].astype("category")

sympols=";/x" #conver all sympol to ','

for sympol in sympols:   
    data['SNP']=data['SNP'].str.replace(sympol,",")
    data['RA']=data['RA'].str.replace(sympol,",")
    data['ChrPos']=data['ChrPos'].str.replace(sympol,",")
    data['ChrId']=data['ChrId'].str.replace(sympol,",")

data['ChrId'] = data['ChrId'].fillna('null')
data['ChrPos'] = data['ChrPos'].fillna('null')

def explode(df, lst_cols, fill_value='', preserve_index=False):
    # make sure `lst_cols` is list-alike
    if (lst_cols is not None
        and len(lst_cols) > 0
        and not isinstance(lst_cols, (list, tuple, np.ndarray, pd.Series))):
        lst_cols = [lst_cols]
    # all columns except `lst_cols`
    idx_cols = df.columns.difference(lst_cols)
    # calculate lengths of lists
    lens = df[lst_cols[0]].str.len()
    # preserve original index values    
    idx = np.repeat(df.index.values, lens)
    # create "exploded" DF
    res = (pd.DataFrame({
                col:np.repeat(df[col].values, lens)
                for col in idx_cols},
                index=idx)
             .assign(**{col:np.concatenate(df.loc[lens>0, col].values)
                            for col in lst_cols}))
    # append those rows that have empty lists
    if (lens == 0).any():
        # at least one list in cells is empty
        res = (res.append(df.loc[lens==0, idx_cols], sort=False)
                  .fillna(fill_value))
    # revert the original index order
    res = res.sort_index()
    # reset index if requested
    if not preserve_index:        
        res = res.reset_index(drop=True)
    return res


data["RA"]=data["RA"].str.split(",")
data["SNP"]=data["SNP"].str.split(",")
data["ChrPos"]=data["ChrPos"].str.split(",")
data["ChrId"]=data["ChrId"].str.split(",")

a=zip(data['RA'],data['SNP'])
b=zip(data['ChrId'],data['SNP'])
c=zip(data['ChrPos'],data['SNP'])

for i,j in a:
    while (len(i)!=len(j)):
        if (len(i)<len(j)):
            i.append("null")
        else:
            j.append("null")

for i,j in b:
    while (len(i)!=len(j)):
        i.append("null")
        
for i,j in c:
    while (len(i)!=len(j)):
        i.append("null")
        
        
data=explode(data,["SNP","ChrPos","ChrId","RA"], fill_value='')

data['ChrPos']=data['ChrPos'].replace(to_replace="null", value = np.nan)
data['ChrId']=data['ChrId'].replace(to_replace="null", value = np.nan)
data['RA']=data['RA'].replace(to_replace="null", value = np.nan)

bool=data['SNP']=="null"
data=data[~bool]

bool1=data["SNP"].astype(str).str.contains("rs")
data=data[bool1]

data=data.reindex(columns=['SNP','RA','pValue','OR_Beta','95%CI','RAF','ChrPos','ChrId','sample','trait_Gwas','annotations','pubmed_id','publication_year','title_pubmedId'])
data
```

```python
#Replace X , Y chromosomes with 23.
data['ChrId']=data['ChrId'].str.replace("X","23").str.replace("Y","23")
```

```python
#Clean the file from weird symbols.  (run requires almost 20 seconds)

import string

rows_for_clean=["SNP","RA","ChrPos","ChrId","trait_Gwas","annotations","pubmed_id","publication_year","title_pubmedId"]

for char in string.punctuation:
    for row in rows_for_clean:
        data[row]=data[row].astype(str).str.replace(char,"")

        
unwanted="+.],['$β€“"
for char in unwanted:
    data["95%CI"]=data["95%CI"].str.replace(char,"")

data    
```

```python
#Separate OR and Beta. The rule is that if the column 95% CI has words near it the number of column OR_Beta is Beta. (except the NR without other word).

data["CI"]=data["95%CI"].copy()  

unwanted="+.],['$β€“- "
for char in unwanted:
    data["CI"]=data["CI"].str.replace(char,"")

b=data["CI"].astype(str).str.isnumeric()
c=data["CI"]=="NR" 
   
bool1 = np.logical_or(b,c)

d=data["OR_Beta"].notnull()
e=data["CI"].isnull()

bool2=np.logical_and(d,e)

bool3=np.logical_or(bool2,bool1)


data["OR"]=data["OR_Beta"][bool3]
data["Beta"]=data["OR_Beta"][~bool3]

data.drop(['CI'], axis=1, inplace=True)

data
```
